```python

def extract_trade_table_with_details(response, base_url="https://www.capitoltrades.com"):
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table')  # Modify selector if needed

    # Initialize lists to store extracted data
    detail_links = []
    tickers = []
    ccys = []
    traded_issuers = []  # If issuer name is available elsewhere

    for row in table.find_all('tr')[1:]:  # Skip header
        # Extract detail link
        link_tag = row.find('a', href=True)
        if link_tag:
            full_url = urljoin(base_url, link_tag['href'])
            detail_links.append(full_url)
        else:
            detail_links.append(None)
        
        # Extract traded issuer details from the 'issuer-ticker' span
        issuer_ticker_span = row.find('span', class_='q-field issuer-ticker')
        if issuer_ticker_span:
            issuer_ticker_text = issuer_ticker_span.get_text(strip=True)
            # Assuming the format is "TICKER:CCY" e.g., "NTAP:US"
            parts = issuer_ticker_text.split(':')
            if len(parts) == 2:
                ticker, ccy = parts
            else:
                ticker, ccy = None, None  # Handle unexpected format
        else:
            ticker, ccy = None, None
        
        tickers.append(ticker)
        ccys.append(ccy)
        
        # If TradedIssuer is available in another cell, extract it here
        # Example:
        # issuer_cell = row.find('td', class_='traded-issuer-name')  # Adjust class name
        # if issuer_cell:
        #     traded_issuer = issuer_cell.get_text(strip=True)
        # else:
        #     traded_issuer = None
        # traded_issuers.append(traded_issuer)
        
        # If TradedIssuer is not available, you can set it same as Ticker or leave it as None
        traded_issuers.append(None)  # Replace None with traded_issuer if available

    # Create DataFrame from the extracted table
    df = pd.read_html(str(table))[0]
    df['Detail_Link'] = detail_links
    df['Ticker'] = tickers
    df['CCY'] = ccys
    df['TradedIssuer'] = traded_issuers  # Add traded issuers if available

    return df
```

```python
import yfinance as yf
import pandas as pd
from datetime import timedelta

def fetch_future_prices(df, date_column='Date', ticker_column='Ticker', days=5):
    """
    Fetches the closing prices of each security for a specified number of days after the trade date.

    Parameters:
    - df (pd.DataFrame): DataFrame containing trade data.
    - date_column (str): Name of the column with trade dates.
    - ticker_column (str): Name of the column with ticker symbols.
    - days (int): Number of days of future prices to fetch.

    Returns:
    - pd.DataFrame: Original DataFrame with additional columns for future prices.
    """
    # Ensure the date column is in datetime format
    df[date_column] = pd.to_datetime(df[date_column])

    # Initialize columns for future prices
    for i in range(1, days + 1):
        df[f'Price_Day_{i}'] = None

    # Get unique tickers to minimize redundant data downloads
    unique_tickers = df[ticker_column].unique()

    # Dictionary to store downloaded data
    ticker_data = {}

    for ticker in unique_tickers:
        try:
            # Download historical data for each ticker
            ticker_history = yf.download(ticker, period='1y', progress=False)
            ticker_history.reset_index(inplace=True)
            ticker_history['Date'] = pd.to_datetime(ticker_history['Date'])
            ticker_data[ticker] = ticker_history
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")
            ticker_data[ticker] = pd.DataFrame()  # Empty DataFrame on failure

    # Iterate over each trade to fetch future prices
    for idx, row in df.iterrows():
        ticker = row[ticker_column]
        trade_date = row[date_column]

        if ticker not in ticker_data or ticker_data[ticker].empty:
            print(f"No data available for ticker: {ticker}")
            continue

        # Filter the ticker history for dates after the trade date
        future_prices = ticker_data[ticker][ticker_data[ticker]['Date'] > trade_date]

        # Get the next 'days' closing prices
        future_prices = future_prices.head(days)

        # Assign the closing prices to the DataFrame
        for i in range(days):
            price = future_prices['Close'].iloc[i] if i < len(future_prices) else None
            df.at[idx, f'Price_Day_{i+1}'] = price

    return df

# Example usage:
# Assuming 'trade_table' is your existing DataFrame with 'Date' and 'Ticker' columns
# trade_table = fetch_future_prices(trade_table, date_column='Date', ticker_column='Ticker', days=5)
# print(trade_table[['Date', 'Ticker', 'Price_Day_1', 'Price_Day_2', 'Price_Day_3', 'Price_Day_4', 'Price_Day_5']])
```

```python
import pandas as pd
from datetime import datetime

def convert_dates(df, columns):
    """
    Converts specified columns in the DataFrame to datetime format.
    Replaces entries with 'today' (case-insensitive) with the current date.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the date columns.
    - columns (list): List of column names to convert.

    Returns:
    - pd.DataFrame: DataFrame with converted date columns.
    """
    for column in columns:
        # Replace 'today' (case-insensitive) with today's date as string
        df[column] = df[column].replace(
            to_replace=r'(?i)^today$',  # Regex to match 'today' case-insensitively
            value=datetime.today().strftime('%Y-%m-%d'),
            regex=True
        )
        
        # Convert the column to datetime
        df[column] = pd.to_datetime(df[column], errors='coerce')
        
    return df

# Example usage:
# Assuming 'trade_table' is your DataFrame and it has 'published' and 'traded' columns
trade_table = convert_dates(trade_table, ['Published', 'Traded'])

# Verify the conversion
print(trade_table[['Published', 'Traded']].dtypes)
print(trade_table[['Published', 'Traded']].head())
```

```python
import yfinance as yf
import pandas as pd

def fetch_historical_prices(df, ticker_column='Ticker', period='1mo'):
    """
    Fetches historical prices for each ticker in the specified DataFrame.

    Parameters:
    - df (pd.DataFrame): DataFrame containing the ticker symbols.
    - ticker_column (str): Name of the column with ticker symbols. Default is 'Ticker'.
    - period (str): Time period for historical data (e.g., '1mo', '3mo', '1y'). Default is '1mo'.

    Returns:
    - dict: A dictionary where keys are ticker symbols and values are DataFrames of historical prices.
    """
    unique_tickers = df[ticker_column].unique()
    historical_data = {}

    for ticker in unique_tickers:
        try:
            data = yf.download(ticker, period=period, progress=False)
            if not data.empty:
                historical_data[ticker] = data
            else:
                print(f"No data found for ticker: {ticker}")
                historical_data[ticker] = pd.DataFrame()
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")
            historical_data[ticker] = pd.DataFrame()

    return historical_data

# Example usage:
# historical_prices = fetch_historical_prices(trade_table, ticker_column='Ticker', period='1mo')
# for ticker, data in historical_prices.items():
#     print(f"Historical data for {ticker}:")
#     print(data.head())
```

```python
def calculate_returns(trade_df, price_df):
    """
    Calculate T+5 and T+10 returns from published date for each trade
    
    Parameters:
    trade_df: DataFrame with at least 'Published' and 'Ticker' columns
    price_df: DataFrame with 'Date', 'Ticker', and 'Close' columns from yfinance
    """
    
    # Ensure dates are datetime
    trade_df['Published'] = pd.to_datetime(trade_df['Published'])
    price_df['Date'] = pd.to_datetime(price_df['date'])
    
    # Initialize lists to store results
    results = []
    
    # Group by ticker to avoid refiltering price data for each trade
    for ticker, trades in trade_df.groupby('Ticker'):
        # Get price data for this ticker
        ticker_prices = price_df[price_df['ticker'] == ticker]
        
        if ticker_prices.empty:
            continue
            
        for _, trade in trades.iterrows():
            pub_date = trade['Published']
            
            # Find the closest trading day on or after published date
            base_price_data = ticker_prices[ticker_prices['Date'] >= pub_date].iloc[0] if not ticker_prices[ticker_prices['Date'] >= pub_date].empty else None
            
            if base_price_data is None:
                continue
                
            # Get base price and date
            base_date = base_price_data['Date']
            base_price = base_price_data['close']
            
            # Find T+5 and T+10 prices
            t5_price = ticker_prices[ticker_prices['Date'] >= (base_date + pd.Timedelta(days=5))]['close'].iloc[0] \
                if not ticker_prices[ticker_prices['Date'] >= (base_date + pd.Timedelta(days=5))].empty else None
                
            t10_price = ticker_prices[ticker_prices['Date'] >= (base_date + pd.Timedelta(days=10))]['close'].iloc[0] \
                if not ticker_prices[ticker_prices['Date'] >= (base_date + pd.Timedelta(days=10))].empty else None
            
            # Calculate returns
            t5_return = ((t5_price - base_price) / base_price * 100) if t5_price is not None else None
            t10_return = ((t10_price - base_price) / base_price * 100) if t10_price is not None else None
            
            results.append({
                'Ticker': ticker,
                'Published_Date': pub_date,
                'Trading_Date': base_date,
                'Base_Price': base_price,
                'T5_Price': t5_price,
                'T10_Price': t10_price,
                'T5_Return_%': round(t5_return, 2) if t5_return is not None else None,
                'T10_Return_%': round(t10_return, 2) if t10_return is not None else None
            })
    
    return pd.DataFrame(results)

# Example usage:
"""
# Assuming you have:
trade_df: DataFrame with columns ['Published', 'Ticker', ...]
price_df: DataFrame with columns ['Date', 'Ticker', 'Close', ...]

# Calculate returns
results = calculate_returns(trade_df, price_df)

# View results
print(results)
"""
```
